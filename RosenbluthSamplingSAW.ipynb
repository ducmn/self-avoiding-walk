```python
from random import seed
from random import random
import math
import numpy as np
import timeit

start = timeit.default_timer()

seed(1)
maxn = 100
np.zeros(maxn)
samples = 2000000
grid = [[False for i in range(2*maxn+1)] for j in range(2*maxn+1)]

def RosenbluthSamplingSAW(x,y,s,w):
    global weight, grid, n
    if s==n:
        weight += w
    else:
        grid[x][y] = True
        avail = []
        if not grid[x][y+1]:
            avail.append(1)
        if not grid[x+1][y]:
            avail.append(2)
        if not grid[x+1][y]:
            avail.append(3)
        if not grid[x+1][y]:
            avail.append(4)
        if len(avail) == 1:
            if avail[0] == 1:
                RosenbluthSamplingSAW(x,y+1,s+1,w*len(avail))
            elif avail[0] == 2:
                RosenbluthSamplingSAW(x+1,y,s+1,w*len(avail))
            elif avail[0] == 3:
                RosenbluthSamplingSAW(x,y-1,s+1,w*len(avail))
            else:
                RosenbluthSamplingSAW(x-1,y,s+1,w*len(avail))
        elif len(avail) == 2:
            roll = random()
            if roll < 0.5:
                if avail[0] == 1:
                    RosenbluthSamplingSAW(x,y+1,s+1,w*len(avail))
                elif avail[0] == 2:
                    RosenbluthSamplingSAW(x+1,y,s+1,w*len(avail))
                elif avail[0] == 3:
                    RosenbluthSamplingSAW(x,y-1,s+1,w*len(avail))
                else:
                    RosenbluthSamplingSAW(x-1,y,s+1,w*len(avail))
            else:
                if avail[1] == 1:
                    RosenbluthSamplingSAW(x,y+1,s+1,w*len(avail))
                elif avail[1] == 2:
                    RosenbluthSamplingSAW(x+1,y,s+1,w*len(avail))
                elif avail[1] == 3:
                    RosenbluthSamplingSAW(x,y-1,s+1,w*len(avail))
                else:
                    RosenbluthSamplingSAW(x-1,y,s+1,w*len(avail))
        elif len(avail) == 3:
            roll = random()
            if roll < 1/3:
                if avail[0] == 1:
                    RosenbluthSamplingSAW(x,y+1,s+1,w*len(avail))
                elif avail[0] == 2:
                    RosenbluthSamplingSAW(x+1,y,s+1,w*len(avail))
                elif avail[0] == 3:
                    RosenbluthSamplingSAW(x,y-1,s+1,w*len(avail))
                else:
                    RosenbluthSamplingSAW(x-1,y,s+1,w*len(avail))
            elif 1/3 <= roll < 2/3:
                if avail[1] == 1:
                    RosenbluthSamplingSAW(x,y+1,s+1,w*len(avail))
                elif avail[1] == 2:
                    RosenbluthSamplingSAW(x+1,y,s+1,w*len(avail))
                elif avail[1] == 3:
                    RosenbluthSamplingSAW(x,y-1,s+1,w*len(avail))
                else:
                    RosenbluthSamplingSAW(x-1,y,s+1,w*len(avail))
            else:
                if avail[2] == 1:
                    RosenbluthSamplingSAW(x,y+1,s+1,w*len(avail))
                elif avail[2] == 2:
                    RosenbluthSamplingSAW(x+1,y,s+1,w*len(avail))
                elif avail[2] == 3:
                    RosenbluthSamplingSAW(x,y-1,s+1,w*len(avail))
                else:
                    RosenbluthSamplingSAW(x-1,y,s+1,w*len(avail))
        elif len(avail) == 4:
            roll = random()
            if roll < 1/4:
                RosenbluthSamplingSAW(x,y+1,s+1,w*len(avail))
            elif 1/4 <= roll < 1/2:
                RosenbluthSamplingSAW(x+1,y,s+1,w*len(avail))
            elif 1/2 <= roll < 3/4:
                RosenbluthSamplingSAW(x,y-1,s+1,w*len(avail))
            else:
                RosenbluthSamplingSAW(x-1,y,s+1,w*len(avail))
    grid[x][y]=False

for n in range(1,maxn+1):
    iti = 0
    weight = 0
    w = 1
    for i in range(1,samples+1):
        RosenbluthSamplingSAW(n,n,0,w)
    esti = weight/samples
    #sd = math.sqrt(4**n * iti/samples * (1-iti/samples))
    print(f'{n} {esti}')
    
stop = timeit.default_timer()

print('Time: ', stop - start) 
```
