```python
from random import seed
from random import random
import math
import timeit

start = timeit.default_timer()

seed(1)
maxn = 100
samples = 1000000
grid=[[False for i in range(2*maxn+1)] for j in range(2*maxn+1)]

def simplesamplingSAW(x,y,s):
    global iti,grid,n,fail
    if s==n:
        iti+=1
    else:
        grid[x][y] = True
        roll = random()
        if 0 <= roll < 0.25:
            if not grid[x][y+1]:
                simplesamplingSAW(x,y+1,s+1)
            else:
                fail += 1
        elif 0.25 <= roll < 0.5:
            if not grid[x+1][y]:
                simplesamplingSAW(x+1,y,s+1)
            else:
                fail += 1
        elif 0.5 <= roll < 0.75:
            if not grid[x][y-1]:
                simplesamplingSAW(x,y-1,s+1)
            else:
                fail += 1
        else:
            if not grid[x-1][y]:
                simplesamplingSAW(x-1,y,s+1)
            else:
                fail += 1
        grid[x][y]=False

for n in range(1,maxn+1):
    iti = 0
    fail = 0
    for i in range(1,samples):
        simplesamplingSAW(n,n,0)
    esti = 4** n * iti/samples
    sd = math.sqrt(4**n * iti/samples * (1-iti/samples))
    print(f'{n} {esti} {sd}')
    
stop = timeit.default_timer()

print('Time: ', stop - start) 
```
